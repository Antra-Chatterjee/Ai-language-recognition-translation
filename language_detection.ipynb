```python
import numpy as np
import pandas as pd

data = pd.read_csv("language.csv")
print(data.head())
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
data.isnull().sum()
```

```python
data['language'].value_counts()
```

```python
data.dtypes
```

```python
x=np.array(data['Text'])
y=np.array(data['language'])
print(y)
```

```python
cv=CountVectorizer()
X=cv.fit_transform(x)
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.33, random_state=42)
X_train
print(X_train)
```

```python
print(y_train)
```

```python
model= MultinomialNB()
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
user= input("Enter a text")
data=cv.transform([user]).toarray()
output=model.predict(data)
print(output)
```

```python
from googletrans import Translator

# 1. User input
user = input("Enter a text: ")

# 2. Detect language with your trained model
data = cv.transform([user]).toarray()
detected_lang = model.predict(data)[0]   # e.g. "French", "Spanish"

print("Detected Language:", detected_lang)

# 3. Ask user for preferred translation target
target_lang = input("Enter target language code (e.g., en for English, fr for French, hi for Hindi): ")

# 4. Translate using googletrans
translator = Translator()
translated = translator.translate(user, dest=target_lang)

print("Translated Text:", translated.text)
```
